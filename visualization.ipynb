```python
%load_ext autoreload
%autoreload 2
#from datasets.dfdc_dataset import DeepFakeTrainingDataset
from datasets.builder import create_dataset
from torch.utils.data import DataLoader
import torchvision
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = create_dataset(dataset='dfdc', data_root_path='/Users/guillermo/Desktop/projects/datasets/dfdc/data_', da_policy='policy_1')
dataset.reset(2,1)
```

```python
sampler = None
loader = DataLoader(dataset, batch_size=4, num_workers=1, shuffle=sampler is None, sampler=sampler, pin_memory=False, drop_last=False)
for i, (data) in enumerate(loader):
    grid = torchvision.utils.make_grid(data["image"], nrow=2)
    fig, ax = plt.subplots(1,1, figsize=(5,5))
    ax.imshow(np.transpose(grid.numpy(), (1,2,0)))
    ax.set_title(f' - face #{data["labels"].sum()}')
    ax.axis('off')
    plt.show()
    #print(data)
    if i > 1:   
        break
```
