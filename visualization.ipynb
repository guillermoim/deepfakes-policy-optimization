Sequence Visualization

```python
from PIL import Image 
import torchvision.transforms as T
import matplotlib.pyplot as plt
import json, torchvision, numpy as np
```

```python
pics = json.load(open('data/faces/chunk_sample/metadata.json'))
to_tensor = T.ToTensor()
```

```python
for video in pics:
    dict_of_pics = pics[video]['faces']
    for key in dict_of_pics:
        
        images = [to_tensor(Image.open(f'data/faces/chunk_sample/{t}')) for t in dict_of_pics[key]]
        grid = torchvision.utils.make_grid(images, nrow=4)
        fig, ax = plt.subplots(1,1, figsize=(10,11))
        ax.imshow(np.transpose(grid.numpy(), (1,2,0)))
        ax.set_title(video)
        ax.axis('off')
        plt.show()
```
